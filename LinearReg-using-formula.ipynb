```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)

X = [499,738,747,808,901,1201,1441,1527,1601,1674,1901,1951,2101,3773]
Y = [17.1,24.85,28,30.22,33,31,53.56,46.5,46,71,68,66,103,126]
mean_x = np.mean(X)
mean_y = np.mean(Y)

n = len(X)
 
# Using the formula to calculate m and c
numer = 0
denom = 0
for i in range(n):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
m = numer / denom
c = mean_y - (m * mean_x)
 
# Print coefficients
print("\n Slope: " ,m,"\nIntercept: ",c)
# Plotting Values and Regression Line
max_x = np.max(X) + 100
min_x = np.min(X) - 100
# Calculating line values x and y
x = np.linspace(min_x, max_x, 1000)
y = c + m * x 
 
# Plotting Line
plt.plot(x, y, color='#52b920', label='Regression Line')
# Plotting Scatter Points
plt.scatter(X, Y, c='#ef4423', label='Scatter Plot')
 
plt.xlabel('Built up area (in sq. ft)')
plt.ylabel('Price (in Lakhs)')
plt.legend()
plt.show()

x=1000
price=m*x+c
print("\n The predicted price for the Built up area ",x," sq.ft is ",price)
```
